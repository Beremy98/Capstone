# Pearson's R recommender 

```python
def pearson(s1, s2): 
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
#Good correlation
pearson(M['adidas Ace 15.1'], M['adidas Ace 16+ Primeknit'])
```

```python
#Good correlation
pearson(M['adidas Copa17.1'], M['Diadora Brasil Italy OG'])
```

```python
#Ok correlation
pearson(M['adidas Predator 18.2'], M['Nike Phantom Venom Elite'])
```

```python
#Bad correlation
pearson(M['adidas Nemeziz 19.1'], M['adidas Nemeziz 19+'])
```

```python
#Recommendation function 

def get_rec(boot, M, num):
    print("\n")
    import numpy as np 
    
    
    reviews = []
    for title in M.columns:
        
        if title == boot:
            continue 
        cor = pearson(M[boot], M[title])
        if np.isnan(cor):       
            continue 
        else: 
            reviews.append((title))
            
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    
    return reviews[:num]
```

```python
#Top 5 recommendations for someone who likes the adidas Copa 17.1

rec = get_rec('adidas Copa17.1', M, 5)
rec
```
